### 9. What are the steps that a data scientist needs to follow to include categorical variables such as NBA position in a linear regression model, in which points scored per game is the target variable? 

#### To include categorical variables in a linear regression model, we need to convert the categorical variable into a quantitative variable by using an indicator/dummy variable. A categorical variable that takes on one of two values can typically be simplified as a quantitative variable with values of 0 and 1. For categorical variables that take on m values where m > 2, we can create m - 1 new variables that take on the values of 0 and 1. Example: NHL positions center, defense, goalie. We create two new variables, Z1 and Z2. When center -> Z1 = 1, Z2 = 0. When defense -> Z1 = 0, Z2 = 1. When goalie -> Z1 = 0, Z2 = 0. This way, each possible value is accounted for.

```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf

## 12. a) Considering the Hitters data set.  Using pandas to read the csv file and create a data-frame called hitters
## and removing observations with missing values
hitters = pd.read_csv('Hitters.csv')
hitters = hitters.dropna()

hitters.head()
```

```python
## b) Creating the following variables:

## log_Salary that represents the log of Salary 
## League_0_1 that takes the value of 1 when League = A and 0 otherwise

hitters['log_Salary'] = np.log(hitters['Salary'])
hitters['League_0_1'] = np.where(hitters['League'] == 'A', 1, 0)

hitters.head()
```

```python
## c) Building a linear regression model in which AtBat, Hits, HmRun, and League_0_1 are the predictor 
## variables and log_Salary is the target variable

## Building the model
lm_md = smf.ols(formula = 'log_Salary ~ AtBat + Hits + HmRun + League_0_1', data = hitters).fit()

## Extracting model results
lm_md.summary()
```

```python
## Using the multiple linear regression model to answer the following:

## i) Reporting the coefficient of determination

## The coefficient of determination (R-squared) value for this model is 0.233.
## In the context of this problem, we can say that 23.3% of the variability in log Salary can be explained 
## by a linear model in which AtBat, Hits, HmRun, and League_0_1 are the predictor variables.
```

```python
## ii) Describing the relationship between log_Salary and AtBat

## Atbat has an estimated coefficient of -0.0022 with an associated standard error of 0.001. Since the estimated coefficient 
## is negative, we know that there is a negative relationship between the two variables. Also, since the estimated coefficient 
## is greater than the margin of error (2 * std. error) in magnitude, we conclude this relationship is significant.
```

```python
## iii) Describing the relationship between log_Salary and League_0_1 

## League_0_1 has an estimated coefficient of -0.1570 with an associated standard error of 0.1. Since the estimated coefficient 
## is negative, we know that there is a negative relationship between the two variables. Also, since the estimated coefficient 
## is less than the margin of error (2 * std. error) in magnitude, we conclude this relationship is not significant.
```

```python
## iv) Predicting the salary of a baseball player in the 1987 season with AtBat = 650, Hits = 250, 
## HmRun = 37 and League_0_1 = 0

## Defining the new data
new_data = pd.DataFrame({'AtBat':[650], 'Hits': [250], 'HmRun': [37], 'League_0_1': [0]})

## Predicting log_Salary
log_salary = lm_md.predict(new_data)

## Predicting the Salary
np.exp(log_salary)

## The estimated salary for a player in 1987 with AtBat = 650, Hits = 250, HmRun = 37 and League_0_1 = 0 is
## approximately $2,829,637.
```

```python
## v) Predicting the salary of a baseball player in the 1987 season with AtBat = 650, Hits = 250, 
## HmRun = 37 and League_0_1 = 1

## Defining the new data
new_data = pd.DataFrame({'AtBat':[650], 'Hits': [250], 'HmRun': [37], 'League_0_1': [1]})

## Predicting log_Salary
log_salary = lm_md.predict(new_data)

## Predicting the Salary
np.exp(log_salary)

## The estimated salary for a player in 1987 with AtBat = 650, Hits = 250, HmRun = 37 and League_0_1 = 1 is
## approximately $2,418,472.
```

```python
## vi) When comparing the salary between players with similar statistics in the American League versus the 
## National League, we can conclude that players from the National League (League_0_1 = 0) have a higher salary, 
## on average, compared to a player from the American League (League_0_1 = 1).
```

```python
## vii) Calculating the Standardized Regression Coefficients

std_log_Salary = hitters['log_Salary'].std()
std_AtBat = hitters['AtBat'].std()
std_Hits = hitters['Hits'].std()
std_HmRun = hitters['HmRun'].std()
std_league_0_1 = hitters['League_0_1'].std()

print('Standardized Coefficient of AtBat:', 0.0022 * std_AtBat / std_log_Salary)
print('Standardized Coefficient of Hits:', 0.0142 * std_Hits / std_log_Salary)
print('Standardized Coefficient of HmRun:', 0.0183 * std_HmRun / std_log_Salary)
print('Standardized Coefficient of League_0_1:', 0.157 * std_league_0_1 / std_log_Salary)

## In terms of the standardized regression coefficient, Hits is the most important variable (highest standardized 
## regression coefficient) and League_0_1 is the least important variable (lowest standardized regression coefficient).
```

```python
## vii) Calculating the Contribution to R-squared values

corr_ls_AtBat = hitters[['log_Salary', 'AtBat']].corr().iloc[0, 1]
corr_ls_Hits = hitters[['log_Salary', 'Hits']].corr().iloc[0, 1]
corr_ls_HmRun = hitters[['log_Salary', 'HmRun']].corr().iloc[0, 1]
corr_ls_League_0_1 = hitters[['log_Salary', 'League_0_1']].corr().iloc[0, 1]


print('The contribution of AtBat to R^2 is', round(100 * 0.0022 * std_AtBat * corr_ls_AtBat / std_log_Salary, 2))
print('The contribution of Hits to R^2 is', round(100 * 0.0142 * std_Hits * corr_ls_Hits / std_log_Salary, 2))
print('The contribution of HmRun to R^2 is', round(100 * 0.0183 * std_HmRun * corr_ls_HmRun / std_log_Salary, 2))
print('The contribution of League_0_1 to R^2 is', round(100 * 0.157 * std_league_0_1 * corr_ls_League_0_1 / std_log_Salary, 2))

## In terms of the contribution of the variable to R2, Hits is the most important variable (highest contribution to R^2) 
## and League_0_1 is the least important variable (lowest contribution to R^2).
```
