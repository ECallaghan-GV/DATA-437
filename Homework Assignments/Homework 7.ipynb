```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf

## 5. a) Using pandas to read the csv file and create a data-frame called hitters and remove observations with missing values

hitters = pd.read_csv('Hitters.csv')
hitters.head()
```

```python
## b) Creating two new variables:
## log_Salary that represents the log of Salary
## League_0_1 that takes the value of 1 when League = A and 0 otherwise

hitters['log_Salary'] = np.log(hitters['Salary'])
hitters['League_0_1'] = np.where(hitters['League'] == 'A', 1, 0)
```

```python
## c) Building a linear regression model in which AtBat, Hits, HmRun, and League_0_1 are the predictor 
## variables and log_Salary is the target variable

## Building the model
mlr_md = smf.ols(formula = 'log_Salary ~ AtBat + Hits + HmRun + League_0_1', data = hitters).fit()

## i) Reporting the coefficient of determination
mlr_md.summary()

## The coefficient of determination (R-squared value) for the given model is 0.233. In terms of this probelm, 
## a coefficient of determination equal to 0.233 means that about 23% of the variability in Y (log_Salary) can 
## be explained by a linear model in which AtBat, Hits, HmRun, and League_0_1 are the predictor variables.
```

```python
## ii) Describing the relationship between log_Salary and AtBat

## log_Salary and AtBat have a slightly negative relationship. The coefficient associated with AtBats is -0.0022 
## with a standard error of 0.001. Since the magnitude of the coefficient is larger than two times the magnitude
## of the standard error, we know that the relationship between log_Salary and AtBat is significant.
```

```python
## iii) Describing the relationship between log_Salary and League_0_1

## The coefficient associated with League_0_1 is -0.1570 with a standard error of 0.1. Since two times the magnitude 
## of the standard error is greater than the magnitude of the coefficient, we know that the relationship between 
## log_Salary and League_0_1 is not statistically significant. 
```

```python
## iv) Predicting the salary of a baseball player in the 1987 season with AtBat = 600, Hits = 220, 
## HmRun = 35 and League_0_1 = 0

## Building the new data frame
data = pd.DataFrame({'AtBat': [600], 'Hits': [220], 'HmRun': [35], 'League_0_1': [0]})

## Predicting on the new data
pred = mlr_md.predict(data)

## Transforming the log prediction into the actual predicted salary
np.exp(pred)

## The estimated salary for a player with AtBat = 600, Hits = 220, HmRun = 35 and League_0_1 = 0
## is approximately $1,990,548.
```

```python
## v) Predicting the salary of a baseball player in the 1987 season with AtBat = 600, Hits = 220,
## HmRun = 35 and League_0_1 = 1.

## Building the new data frame
data = pd.DataFrame({'AtBat': [600], 'Hits': [220], 'HmRun': [35], 'League_0_1': [1]})

## Predicting on the new data
pred = mlr_md.predict(data)

## Transforming the log prediction into the actual predicted salary
np.exp(pred)

## The estimated salary for a player with AtBat = 600, Hits = 220, HmRun = 35 and League_0_1 = 1
## is approximately $1,701,308.
```
