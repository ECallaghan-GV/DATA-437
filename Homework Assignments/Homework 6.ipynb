### 4. If the coefficient of determination of a simple linear regression model is 97.5%, what can you tell about the slope of the line? That is, is it positive or negative? 

#### If the coefficient of determination of a simple linear regression model is 97.5%, we can conclude that there is a strong linear relationship between the two variables. R-squared is a measure of how well a linear relationship fits the data, and the values range from 0 to 1. Therefore, we cannot determine if the slope of the line is positive or negative, only how well it is fit by a linear function.

```python
## 5. a) Using pandas to read the csv file and create a data-frame called mlb

import pandas as pd
import numpy as np
import statsmodels.formula.api as smf

mlb = pd.read_csv('Teams.csv')
mlb.head()
```

```python
## b) Creating two new variables: RD (run differential as R - RA) and Wpct (winning percentage as W / (W + L))

mlb['RD'] = mlb['R'] - mlb['RA']
mlb['Wpct'] = mlb['W'] / (mlb['W'] + mlb['L'])
```

```python
## c) We are interested in studying the relationship between RD and Wpct for recent seasons. 
## Subsetting the data on seasons since 2001

mlb_subset = mlb[mlb['yearID'] >= 2001]
mlb_subset.head()
```

```python
## d) Building a simple linear regression model in which RD is the predictor variable and
## Wpct is the target variable

## Building the model
lm_md = smf.ols(formula = 'Wpct ~ RD', data = mlb_subset).fit()

## Defining the new data
new_data = pd.DataFrame({'RD': [87]})

## Predicting the Wpct when RD = 87
lm_md.predict(new_data)

## The predicted Winning Percentage of a team with a Run Differential of 87 is
## approximately 55.5%.
```

```python
## e) The slope of the model from part (d) is statistically significant because the magnitude of the
## slope, 0.0006, is greater than two times the magnitude of the standard error associated with the slope.

lm_md.summary()

## Slope = 0.0006, Standard Error = 0.0000106
```
