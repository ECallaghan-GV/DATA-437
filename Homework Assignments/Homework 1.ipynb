```python
## 3. a) Using the pandas library to read the csv file and create a data-frame called ufc

import pandas as pd

ufc = pd.read_csv('UFC_data.csv')
ufc.head()
```

```python
## b) Removing all the observations in which weight class is equal to OpenWeight

ufc = ufc[ufc['weight_class'] != 'OpenWeight']
```

```python
## c) Reporting the corner with highest winning rate

winners = ufc.groupby('Winner')[['R_fighter', 'B_fighter']].count()

winners['winning_rate'] = (winners['R_fighter'] / winners['R_fighter'].sum()).round(3) * 100

winners

## The red corner has the highest winning rate at about 66% of fights.
```

```python
## d) Reporting the corner with highest winning rate across the different weight classes

winners = ufc.groupby(['weight_class', 'Winner'])[['R_fighter', 'B_fighter']].count()

winners['winning_rate'] = (winners['R_fighter'] / winners['R_fighter'].groupby('weight_class').sum()).round(3) * 100

winners

## For every weight class, the red corner has the higher winning rate (over 50% for each class). This is consistent with the answer from part (c)
## in which we found that the red corner has the highest winning rate independent on the weight class.
```

```python
## e) Reporting the corner with highest winning rate across the different weight classes and whether or not the fight is a title bout

winners = ufc.groupby(['title_bout', 'weight_class', 'Winner'])[['R_fighter', 'B_fighter']].count()

winners['winning_rate'] = (winners['R_fighter'] / winners['R_fighter'].groupby(['title_bout', 'weight_class']).sum()).round(3) * 100

pd.set_option('display.max_rows', None, 'display.max_columns', None)

winners

## Whether or not the fight is a title bout, the red corner shows the highest winning rate across all different weight classes. This is consistent 
## with the answers from parts (c)-(d) in which we found that for all fights, the red corner shows the highest winning rate (across all fights and 
## all weight classes)
```
