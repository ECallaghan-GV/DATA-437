```python
## 1. a) Using pandas to read the csv file and create a data-frame called teams

import pandas as pd
import numpy as np
import statsmodels.formula.api as smf

teams = pd.read_csv('Teams.csv')
teams.head()
```

```python
## b) Subsetting the data to only include:
## G, W, L, R, RA, teamID, yearID, and lgID

my_teams = teams[['teamID', 'yearID', 'lgID', 'G', 'W', 'L', 'R', 'RA']]
my_teams.head()
```

```python
## c) Computing the runs differential and winning percentage 

my_teams['RD'] = my_teams['R'] - my_teams['RA']
my_teams['Wpct'] = my_teams['W'] / (my_teams['W'] + my_teams['L'])
```

```python
## d) Refitting the linear model in which RD is the predictor variable and Wpct is the target
## variable for the seasons 1961-1970, the seasons 1971-1980, the seasons 1981-1990, the
## seasons 1991-2000, and the seasons 2001-2010.

## 60's
my_teams_60s = my_teams[(my_teams['yearID'] > 1960) & (my_teams['yearID'] <= 1970)]
lm_md_60 = smf.ols(formula = 'Wpct ~ RD', data = my_teams_60s).fit()

## 70's
my_teams_70s = my_teams[(my_teams['yearID'] > 1970) & (my_teams['yearID'] <= 1980)]
lm_md_70 = smf.ols(formula = 'Wpct ~ RD', data = my_teams_70s).fit()

## 80's
my_teams_80s = my_teams[(my_teams['yearID'] > 1980) & (my_teams['yearID'] <= 1990)]
lm_md_80 = smf.ols(formula = 'Wpct ~ RD', data = my_teams_80s).fit()

## 90's
my_teams_90s = my_teams[(my_teams['yearID'] > 1990) & (my_teams['yearID'] <= 2000)]
lm_md_90 = smf.ols(formula = 'Wpct ~ RD', data = my_teams_90s).fit()

## 00's
my_teams_00s = my_teams[(my_teams['yearID'] > 2000) & (my_teams['yearID'] <= 2010)]
lm_md_00 = smf.ols(formula = 'Wpct ~ RD', data = my_teams_00s).fit()
```

```python
## e) Comparing across the five decades the predicted winning percentage for a team with a run differential of 10 runs

## Defining a new data point
new_data = pd.DataFrame({'RD': [10]})

## Defining a list of models
models = [lm_md_60, lm_md_70, lm_md_80, lm_md_90, lm_md_00]
model_numbers = ["60's", "70's", "80's", "90's", "00's"]

for i in range(0, len(models)):
    print(model_numbers[i], 'Wpct Model for a team with RD = 10:', round(models[i].predict(new_data)[0] * 100, 3), '%')


## Between the decades, the values for winning percentage are changing very slightly. The maximum is in the 60's,
## where a team with RD = 10 is expected to win about 50.7% of games played. The minimum is in the 2000's, where
## a team with RD = 10 is expected to win about 50.62% of games played.
```
